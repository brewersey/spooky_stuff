```python
#Create Sentiment Analyzer
analyzer = SentimentIntensityAnalyzer()
sentiments = []

for sentence in df_authors.text:
    feeling = analyzer.polarity_scores(sentence)
    sentiments.append(feeling)
```

```python
stopWords = set(stopwords.words('english'))
```

```python
authors_clf = Pipeline([('vect', CountVectorizer(stop_words=stopWords)),
                        ('tfidf', TfidfTransformer()),
                        ('clf', SGDClassifier(loss='hinge', penalty='l2',
                                           alpha=1e-3, n_iter=5, random_state=42)),
])
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_authors.text, df_authors.author, test_size = 0.3)
```

```python
y_train.value_counts()
```

```python
authors_clf.fit(X_train, y_train)
```

```python
predicted = authors_clf.predict(X_train)
```

```python
np.mean(predicted == y_train) 
```

```python
#set grid search parameters
parameters = {'vect__ngram_range': [(1, 1), (1, 2), (1, 3)],
              'tfidf__use_idf': (True, False),
              'clf__alpha': (1e-2, 1e-3),
}
```

```python
gs_clf = GridSearchCV(authors_clf, parameters, n_jobs=-1)
```

```python
gs_clf = gs_clf.fit(X_train, y_train)
predictions = gs_clf.predict(X_train)
```

```python
np.mean(predictions == y_train)
```

```python
cm = metrics.confusion_matrix(y_train, predictions, ['HPL', 'EAP', 'MWS'])
print(cm)
```

```python
#nice plot of confusion matrix; class bias becomes clearly apparent
labels = ['HPL', 'EAP', 'MWS']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cm)
plt.title('Confusion Matrix for Author Identification')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
#importing Keras Libaries
import keras
from keras.models import Sequential, Model
from keras import layers
from keras.layers import Input, Dropout, Permute, Dense, Flatten
from keras import optimizers
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import normalize
from keras.utils import to_categorical
```

```python
df_authors.head()
```

```python
stopWords = set(stopwords.words('english'))
tfidf_vec = TfidfVectorizer(ngram_range=(1, 2), 
                            stop_words=stopWords)
```

```python
word_vec_nn = tfidf_vec.fit_transform(df_authors.text)
```

```python
word_vec_nn.shape
```

```python
pd.DataFrame(word_vec_nn.todense(), 
             columns=tfidf_vec.get_feature_names()
            ).head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(word_vec_nn, df_authors.author, test_size = 0.3)
```

```python
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(y_train)
encoded_y_train = encoder.transform(y_train)

encoder = LabelEncoder()
encoder.fit(y_test)
encoded_y_test = encoder.transform(y_test)

# convert integers to dummy variables (i.e. one hot encoded)
y_train = to_categorical(encoded_y_train, num_classes = 3)
y_test = to_categorical(encoded_y_test, num_classes = 3)
```

```python
X_train = X_train.todense()
X_test = X_test.todense()
```

```python
n_cols = X_train.shape[1]
```

```python
model = Sequential()
```

```python
model.add(Dense(5, activation='relu', input_shape = (n_cols, )))
model.add(Dropout(0.15))
#model.add(Dense(50, activation='relu'))
#model.add(Dropout(0.15))
#model.add(Dense(50, activation='relu'))
#model.add(Dropout(0.15))
model.add(Dense(3, activation='softmax'))
```

```python
sgd = optimizers.SGD(lr=0.001, decay=1e-6, momentum=0.9, nesterov=True)
adx = optimizers.Adamax(lr=0.002, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)

model.compile(optimizer= 'adam', 
              loss= 'categorical_crossentropy',
              metrics= ['accuracy'])
```

```python
model.fit(X_train, y_train,
          validation_split = 0.3,
          #batch_size = 128,
          epochs = 5)
```

```python
evaluations = model.evaluate(X_test, y_test)
print('loss: ' + str(evaluations[0]))
print('acc: ' + str(evaluations[1]))
```
